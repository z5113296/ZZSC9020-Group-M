# Australia Population 

### Data Source

Australian Bureau of Statistics
https://www.abs.gov.au/statistics/people/population/national-state-and-territory-population/latest-release

```python
# Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load Population Data
sheet_name='Data1'
pop_raw=pd.read_excel('~/12. ZZSC 9020 Data Science Project/Project/Data/Population/All Time Series Spreadsheets/National State and Territory Population.xlsx', sheet_name=sheet_name)
```

```python
pop_raw.columns.tolist()
```

```python
pop=pop_raw[['Unnamed: 0','Estimated Resident Population (ERP) ;  Australia ;']][9:]
```

```python
#rename columns
```

```python
pop=pop.rename(columns={'Unnamed: 0':'Year_Quarter','Estimated Resident Population (ERP) ;  Australia ;':'Population in Thousands'})
```

```python
pop=pop.reset_index(drop=True)
```

```python
pop
```

```python
# Convert Year_Quarter column to datetime format
pop['Year_Quarter'] = pd.to_datetime(pop['Year_Quarter'])
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x=pop['Year_Quarter'], y=pop['Population in Thousands'], marker='o')
plt.title('Population Trend Over Time')
plt.xlabel('Year and Quarter')
plt.ylabel('Population in Thousands')
plt.show()
```

# Australia Fuel Prices

### Data Source

Australian Government - Department of Climate Change, Energy, the Environment and Water
https://www.energy.gov.au/publications/australian-petroleum-statistics-2024

```python
# Load Data
sheet_name='Australian fuel prices'
fuel_raw=pd.read_excel('~/12. ZZSC 9020 Data Science Project/Project/Data/Fuel Price/Australian Petroleum Statistics - Data Extract January 2024.xlsx', sheet_name=sheet_name)
```

```python
fuel_raw['Year_Quarter']=fuel_raw['Year'].astype(str)+fuel_raw['Quarter'].astype(str)
```

```python
fuel_raw.columns.tolist()
```

```python
fuel_raw
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(rotation='vertical')
plt.title('Premium Unleaded Petrol Prices Trend Over Time')
plt.xlabel('Year and Quarter')
plt.ylabel('Price (cents per litre)')
sns.lineplot(x='Year_Quarter', y='Premium unleaded petrol (95 RON) (cpl)', data=fuel_raw, marker='o',color='red',label='Premium')
sns.lineplot(x='Year_Quarter', y='Regular unleaded petrol (91 RON) (cpl)', data=fuel_raw, marker='o',color='blue',label='Regular')
sns.lineplot(x='Year_Quarter', y='Automotive diesel (cpl)', data=fuel_raw,marker='o',color='green', label='Diesel')
sns.lineplot(x='Year_Quarter', y='Automotive LPG (cpl)', data=fuel_raw, marker='o',color='yellow', label='LPG')  # should be removed from analysis. This is liquid propane
plt.show()
```

```python
# Find Price Outlier for Premium unleaded petrol (95 RON)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Plot the line plot with markers
sns.lineplot(x='Year_Quarter', y='Premium unleaded petrol (95 RON) (cpl)', data=fuel_raw, marker='o', color='red', label='Premium', ax=axes[0])
axes[0].set_title('Line Plot')
axes[0].set_xlabel('Year and Quarter')
axes[0].set_ylabel('Price (cents per litre)')
axes[0].legend()

# Plot the boxplot
sns.boxplot(y='Premium unleaded petrol (95 RON) (cpl)', data=fuel_raw, color='blue', ax=axes[1])
axes[1].set_title('Boxplot')
axes[1].set_ylabel('Price (cents per litre)')

# Show the plots
plt.tight_layout()  # Adjusts subplot parameters to give specified padding
plt.show()
```

# Coal Price 

### Data Source

Federal Reserve Bank of St Louis
https://fred.stlouisfed.org/series/PCOALAUUSDM

```python
# load data
coal_raw=pd.read_csv('~/12. ZZSC 9020 Data Science Project/Project/Data/Coal Price/PCOALAUUSDM.csv')
```

```python
coal_raw.set_index('DATE',inplace=True)
coal_raw.index=pd.to_datetime(coal_raw.index)
```

```python
# aggregate monthly data at quarterly basis
coal=coal_raw.resample('Q').mean()
```

```python
coal
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(rotation='vertical')
sns.lineplot(data=coal, x='DATE', y='PCOALAUUSDM')
```

# Australia Electricity Price

### Data Source

Australian Government - Australia Energy Regulator
https://www.aer.gov.au/industry/registers/charts/annual-volume-weighted-average-30-minute-prices-regions

```python
# load data
ele_raw=pd.read_csv('~/12. ZZSC 9020 Data Science Project/Project/Data/Australia Electricity Price/Electricity_Price.csv')
```

```python
ele_raw.columns.tolist()
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(rotation='vertical')
melted_df = ele_raw.melt(id_vars=['Financial year'], var_name='Column', value_name='Value')
sns.lineplot(data=melted_df, x='Financial year', y='Value', hue='Column', marker='o')
plt.title('Electricty Price Over Years')
plt.xlabel('Financial Years')
plt.ylabel('Electricity Price')
plt.legend(title='State')  
```
