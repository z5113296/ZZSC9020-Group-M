```python
# Import library
%pip install seaborn
%pip install openpyxl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load Population Data
df = pd.read_excel('./GDP_Australia.xlsx')
df.columns.tolist()
```

```python
#Rename columns
df.columns = ['Time_Period', 'GDP', 'GDP_Per_Capita']
df.rename(columns={'Time Period': 'Time_Period',
                   'Gross domestic product (AUD, Millions)': 'GDP',
                   'GDP/Capita': 'GDP_Per_Capita'}, inplace=True)
```

```python
df.head()
```

```python
#Split years and quarters for graphing
df[['Year', 'Quarter']] = df['Time_Period'].str.split('-', expand=True)
df['Year'] = df['Year'].astype(int) 
df['Quarter'] = df['Quarter'].astype('category')
df.head()
```

```python
# Plotting GDP over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Time_Period', y='GDP', data=df, marker='o', hue='Year', palette='viridis')
plt.title('GDP Over Time')
plt.xlabel('Time Period')
plt.ylabel('GDP (Millions)')
plt.xticks(rotation=90)
plt.tight_layout()  
plt.show()
```

```python
# Plotting GDP Per Capita over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Time_Period', y='GDP_Per_Capita', data=df, marker='o', hue='Year', palette='viridis')
plt.title('GDP Per Capita Over Time')
plt.xlabel('Time Period')
plt.ylabel('GDP Per Capita (Millions)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Creating boxplots to see the distribution of GDP by Quarter
plt.figure(figsize=(12, 6))
sns.boxplot(x='Quarter', y='GDP', data=df)
plt.title('GDP Distribution by Quarter')
plt.xlabel('Quarter')
plt.ylabel('GDP (Millions)')
plt.show()
```
